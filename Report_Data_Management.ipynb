```python
def feature_selection(dataframe, feature_list):
  '''
    This is a method which returns the new dataframe in which the features are selected
  '''
  return dataframe[feature_list]
```

```python
# Fetch Raw Data -> Select necessary features by BQ's request -> Save the datasets of Company Variables and Companies to CSVs


company_variables_ts_df1 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_1.csv'), company_variables_ts_and_features)
company_variables_ts_df1.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_1.csv')

company_variables_ts_df2 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_2.csv'), company_variables_ts_and_features)
company_variables_ts_df2.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_2.csv')

company_variables_ts_df3 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_3.csv'), company_variables_ts_and_features)
company_variables_ts_df3.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_3.csv')

company_variables_ts_df4 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_4.csv'), company_variables_ts_and_features)
company_variables_ts_df4.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_4.csv')

company_variables_ts_df5 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_5.csv'), company_variables_ts_and_features)
company_variables_ts_df5.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_5.csv')

company_variables_ts_df6 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_6.csv'), company_variables_ts_and_features)
company_variables_ts_df6.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_6.csv')

company_variables_ts_df7 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/company_variables_ts_7.csv'), company_variables_ts_and_features)
company_variables_ts_df7.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_7.csv')

companies_df1 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/companies_1.csv'), company_info)
companies_df1.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_1.csv')

companies_df2 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/companies_2.csv'), company_info)
companies_df2.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_2.csv')

companies_df3 = feature_selection(pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data/companies_3.csv'), company_info)
companies_df3.to_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_3.csv')
```

```python
#Read the company variables datasets and select 2015 - 2020


company_variables_ts_df1 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_1.csv')
company_variables_ts_df2 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_2.csv')
company_variables_ts_df3 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_3.csv')
company_variables_ts_df4 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_4.csv')
company_variables_ts_df5 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_5.csv')
company_variables_ts_df6 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_6.csv')
company_variables_ts_df7 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/company_variables_ts_7.csv')

company_variables_ts_df1 = company_variables_ts_df1[(company_variables_ts_df1.bq_year >= 2015) & (company_variables_ts_df1.bq_year <= 2020)]
company_variables_ts_df2 = company_variables_ts_df2[(company_variables_ts_df2.bq_year >= 2015) & (company_variables_ts_df2.bq_year <= 2020)]
company_variables_ts_df3 = company_variables_ts_df3[(company_variables_ts_df3.bq_year >= 2015) & (company_variables_ts_df3.bq_year <= 2020)]
company_variables_ts_df4 = company_variables_ts_df4[(company_variables_ts_df4.bq_year >= 2015) & (company_variables_ts_df4.bq_year <= 2020)]
company_variables_ts_df5 = company_variables_ts_df5[(company_variables_ts_df5.bq_year >= 2015) & (company_variables_ts_df5.bq_year <= 2020)]
company_variables_ts_df6 = company_variables_ts_df6[(company_variables_ts_df6.bq_year >= 2015) & (company_variables_ts_df6.bq_year <= 2020)]
company_variables_ts_df7 = company_variables_ts_df7[(company_variables_ts_df7.bq_year >= 2015) & (company_variables_ts_df7.bq_year <= 2020)]
```

```python
#Read the company datasets and select Public Companies


companies_df1 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_1.csv')
companies_df2 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_2.csv')
companies_df3 = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/cleaned_data_state/companies_3.csv')

companies_df1 = companies_df1[(companies_df1.bq_public_indicator == 1)]
companies_df2 = companies_df2[(companies_df2.bq_public_indicator == 1)]
companies_df3 = companies_df3[(companies_df3.bq_public_indicator == 1)]
```

```python
# Combine the 7 company_variables_ts_df dataframes together, Remove new columns that BQ wants to exclude and drop NANs


company_variables_ts_df = pd.concat([company_variables_ts_df1, company_variables_ts_df2, company_variables_ts_df3, company_variables_ts_df4, 
                company_variables_ts_df5, company_variables_ts_df6, company_variables_ts_df7], ignore_index=True)

company_variables_ts_df = company_variables_ts_df.drop([company_variables_ts_df.columns[0]], axis = 1)
company_variables_ts_df.drop("bq_fiscal_year_end", inplace = True, axis = 1)
company_variables_ts_df = company_variables_ts_df.dropna()
```

```python
# Combine the 3 company_df dataframes together, Remove new columns that BQ wants to exclude and drop NANs


companies_df = pd.concat([companies_df1, companies_df2, companies_df3], ignore_index=True)
companies_df = companies_df.drop([companies_df.columns[0]], axis = 1)
companies_df = companies_df.dropna()
```

```python
# Save the new processed Companies and Company Variables Datasets


companies_df.to_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/companies_df.csv')
company_variables_ts_df.to_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/company_variables_ts_df.csv')
```

```python
# Read the saved Companies and Company Variables Datasets, drop unnamed column


companies_df = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/companies_df.csv')
companies_df = companies_df.drop([companies_df.columns[0]], axis = 1)

company_variables_ts_df = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/company_variables_ts_df.csv')
company_variables_ts_df = company_variables_ts_df.drop([company_variables_ts_df.columns[0]], axis = 1)
```

```python
# Merge the processed Companies Table and Company Variables Table by company bq id


companies_company_variables = pd.merge(company_variables_ts_df, companies_df, on = "bq_id")
```

```python
# Read Statistics Raw Dataset and Table of Statistic fields that BQ wants to investigate, Select the Stats data from 2015-2020
# If stat field's value = NAN then replace it with 0


stats_by_us_ts = pd.read_csv("/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/statistics_by_us_ts.csv")

meta_features = pd.read_csv("/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/stats_fields")
meta_features = meta_features.drop([meta_features.columns[0]], axis=1) 

selected_meta = ["bq_year"]
selected_meta.extend(list(meta_features["Field name"]))
stats_by_us_ts = stats_by_us_ts[(stats_by_us_ts.bq_year >= 2015) & (stats_by_us_ts.bq_year <= 2020)][selected_meta].fillna(0)
```

```python
# Save the processed Stats Table


stats_by_us_ts.to_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/stats_fillna.csv')
```

```python
# Merge the processed Stats Table and Companies + Company Variables Table by year


stats_by_us_ts = pd.read_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/stats_fillna.csv')

merged_meta_data = pd.merge(companies_company_variables, stats_by_us_ts, on = "bq_year")
```

```python
# Save the merged Companies + Company Variables + Stats as the dataset we want to use


merged_meta_data.sort_values(by=['bq_id','bq_year']).to_csv('/content/drive/MyDrive/BrightQuery_Project/Report_Sample_Data/merged_meta_data.csv')
```
